```python
import os
import matplotlib.pyplot as plt
import random
import cv2
import numpy as np
```

```python
classes = ['field', 'road']
class_index = {'field': 0, 'road': 1}

dataset_path = 'dataset'
field_data_path = 'dataset/fields'
road_data_path = 'dataset/roads'
test_data_path = 'dataset/test_images'
```

```python
class_count = {
    classes[0]: len(os.listdir(field_data_path)),
    classes[1]: len(os.listdir(road_data_path)),
}

print(class_count)
print("Number of image contains in test dataset :    ", len(os.listdir(test_data_path)))
```

```python
plt.bar(class_count.keys(), class_count.values(), width = .5);
plt.title("Number of Images by Class");
plt.xlabel('Class Name');
plt.ylabel('Images count.');
```

```python
def show_random_images(path_to_directory):
    image_files = os.listdir(path_to_directory)
    random.shuffle(image_files)
    fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(10, 10))

    for i, ax in enumerate(axes.flatten()):
        img = cv2.imread(os.path.join(path_to_directory, image_files[i]))
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        ax.imshow(img)
        ax.set_xticks([])
        ax.set_yticks([])
        ax.set_title(image_files[i], fontsize=10)

    plt.tight_layout()
    plt.show()
```

```python
# Fields

show_random_images(path_to_directory=field_data_path)
```

```python
# Roads

show_random_images(path_to_directory=road_data_path)
```

```python
def image_channels(path_to_directory, class_name):
    channels = []

    for filename in os.listdir(path_to_directory):
        img = cv2.imread(os.path.join(path_to_directory, filename))
        if img is not None:
            channels.append(img.shape[2:])
    print(f"Channels contains in {class_name} dataset :    ", set(channels))


image_channels(path_to_directory=field_data_path, class_name="Field")
image_channels(path_to_directory=road_data_path, class_name="Road")
image_channels(path_to_directory=test_data_path, class_name="Test")
```

```python
# Size of image exploring

def explore_image_sizes(path_to_directory, class_name):
    image_sizes = []

    for filename in os.listdir(path_to_directory):
        img = cv2.imread(os.path.join(path_to_directory, filename))
        if img is not None:
            image_sizes.append(img.shape[:2])

    image_sizes = np.array(image_sizes)

    plt.figure(figsize=(8, 6))
    plt.scatter(image_sizes[:, 0], image_sizes[:, 1], alpha=0.5)
    plt.title(f'Image Sizes in {class_name} dataset')
    plt.xlabel('Width')
    plt.ylabel('Height')
    plt.show()

    print('Minimum image size:', np.min(image_sizes, axis=0))
    print('Maximum image size:', np.max(image_sizes, axis=0))
    print('Average image size:', np.mean(image_sizes, axis=0))
    print('Median image size:', np.median(image_sizes, axis=0))
```

```python
explore_image_sizes(path_to_directory=field_data_path, class_name='field')
```

```python
explore_image_sizes(path_to_directory=road_data_path, class_name='road')
```

```python
explore_image_sizes(path_to_directory=test_data_path, class_name='test')
```
